# Comparação - Embeddings

```python
import pandas as pd
import scipy.stats as st
```

```python
embeddings = pd.read_csv("embeddings.csv")
metrics = ["precision","recall","bert_score"]
```

```python
embeddings
```

```python
for metric in metrics:
    print(f"{metric.capitalize()} Mínimo {embeddings[metric].min()}, Máximo {embeddings[metric].max()}")
```

## Média das métricas em cada modelo

```python
confidence_intervals = []
confidence = 0.95

for metric in metrics:
    per_group = embeddings.groupby("model")[metric]
    for row in per_group.agg(["mean","count",st.sem]).iterrows():
        range_min,range_max = st.t.interval(confidence=confidence,df=row[1]["count"]-1,loc=row[1]["mean"],scale=row[1]["sem"])
        confidence_intervals.append({"metric":metric,"model":row[0], "range_min":range_min,"mean": row[1]["mean"], "range_max":range_max})
```

```python
interval_df = pd.DataFrame(confidence_intervals)
interval_df
```

```python
interval_df.to_csv("confidence_intervals_embeddings.csv")
```
